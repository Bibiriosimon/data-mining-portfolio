```python
#数据加载与初步分析
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体，防止乱码
plt.rcParams['font.sans-serif'] = ['Heiti TC']      # 或 ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False        # 解决负号显示问题

#描述性统计
data = pd.read_csv('../datasets/credit_risk_data.csv')
print(data.head())
print(data.info())
print(data.describe())
print(data.isnull().sum())

#绘制数值特征的分布直方图
numerical_features = ['age', 'income', 'loan_amount', 'credit_score', 'employment_years']

for col in numerical_features:
    plt.figure(figsize=(8, 5)) 
    sns.histplot(data=data, x=col, kde=True, bins=20)
    plt.title(f'"{col}" 的分布直方图', fontsize=16)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('频数', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

risk_counts = data['risk'].value_counts()
print(risk_counts)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

processed_data = data.copy()

label_encoder = LabelEncoder()
processed_data['loan_purpose'] = label_encoder.fit_transform(processed_data['loan_purpose'])

print("--- 预处理后的数据 (前5行) ---")
print(processed_data.head())
print("\n'loan_purpose' 列的数据类型:", processed_data['loan_purpose'].dtype)

X = processed_data.drop('risk', axis=1)
y = processed_data['risk']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,      
    random_state=42,    
    stratify=y          
)

print(f"\n训练集形状: {X_train.shape}, {y_train.shape}")
print(f"测试集形状: {X_test.shape}, {y_test.shape}")

numerical_cols = ['age', 'income', 'loan_amount', 'credit_score', 'employment_years']
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

print("\n--- 标准化后的训练集 (前5行) ---")
print(X_train.head())
print("\n--- 标准化后的测试集 (前5行) ---")
print(X_test.head())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score


# --- 嵌套交叉验证配置 ---
# 创建一个包含数据缩放和模型训练的 Pipeline
# 这样做可以防止数据泄露，因为在交叉验证的每一折中，StandardScaler都只会在训练数据上fit
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', DecisionTreeClassifier(random_state=42))
])

# 定义决策树的参数网格
param_grid = {
    'model__criterion': ['gini', 'entropy'],
    'model__max_depth': [3, 5, 7, 10, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 3, 5]
}

# 内层交叉验证（用于GridSearchCV调参）
inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 外层交叉验证（用于评估最终性能）
outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

# 将网格搜索(GridSearchCV)封装起来，它将作为我们要在外层CV中评估的“模型”
gs = GridSearchCV(
    estimator=pipeline, 
    param_grid=param_grid, 
    cv=inner_cv, 
    scoring='f1', # 对于二分类，'f1' 等同于正类的f1。如果类别不平衡，'f1_macro' 或 'f1_weighted' 更好
    n_jobs=-1     # 使用所有CPU核心
)

# --- 执行嵌套交叉验证 ---
print("--- 开始执行嵌套交叉验证（手动循环）---\n")
outer_scores = []
best_params_list = []
outer_splits = []

for i, (train_idx, test_idx) in enumerate(outer_cv.split(X, y)):
    outer_splits.append({'train': train_idx, 'test': test_idx})

    # 划分外层训练集和测试集
    X_train_outer, X_test_outer = X.iloc[train_idx], X.iloc[test_idx]
    y_train_outer, y_test_outer = y.iloc[train_idx], y.iloc[test_idx]
    
    # 1. 在外层训练集上运行GridSearchCV (内层循环)
    gs.fit(X_train_outer, y_train_outer)
    
    # 2. 记录内层循环找到的最佳参数
    best_params = gs.best_params_
    best_params_list.append(best_params)
    print(f"本折找到的最佳参数: {best_params}")
    
    # 3. 使用找到的最佳模型对外层测试集进行预测和评估
    # gs.best_estimator_ 已经是使用最佳参数在整个 X_train_outer 上重新训练过的模型
    y_pred = gs.best_estimator_.predict(X_test_outer)
    score = f1_score(y_test_outer, y_pred)
    outer_scores.append(score)
    print(f"本折在外层测试集上的 F1 得分: {score:.4f}\n")

# --- 4. 结果分析 ---
print("\n--- 嵌套交叉验证最终结果 ---")
outer_scores = np.array(outer_scores)
mean_f1 = outer_scores.mean()
std_f1 = outer_scores.std()

print(f"外层交叉验证各折的 F1 得分: {outer_scores}")
print(f"平均 F1 性能: {mean_f1:.4f} ± {std_f1:.4f}")

# 找到 F1 分数最高的那一折及其参数
best_fold_index = np.argmax(outer_scores)
best_f1_overall = outer_scores[best_fold_index]
best_params_overall = best_params_list[best_fold_index]

print(f"\n在所有折中，最高的 F1 分数是: {best_f1_overall:.4f}")
print(f"达到最高分数时的参数组合是: {best_params_overall}")

```

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report, roc_curve, auc
# --- 5. 对表现最好的一折进行详细分析  ---
print("\n\n--- 详细分析表现最好的一折 ---")
# 步骤 5.1: 获取最佳折的数据
best_split = outer_splits[best_fold_index]
best_train_idx, best_test_idx = best_split['train'], best_split['test']
X_train_best_fold, X_test_best_fold = X.iloc[best_train_idx], X.iloc[best_test_idx]
y_train_best_fold, y_test_best_fold = y.iloc[best_train_idx], y.iloc[best_test_idx]
# 步骤 5.2: 创建并用最佳参数训练模型
print("\n正在使用最佳参数重新训练模型...")
# 创建一个新的Pipeline实例
best_model_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', DecisionTreeClassifier(random_state=42))
])
# 设置在最佳折上找到的最佳参数
best_model_pipeline.set_params(**best_params_overall)
# 在最佳折的训练集上训练模型
best_model_pipeline.fit(X_train_best_fold, y_train_best_fold)
print("模型训练完成。")
# 步骤 5.3: 在最佳折的测试集上进行预测
y_pred_best = best_model_pipeline.predict(X_test_best_fold)
y_pred_proba_best = best_model_pipeline.predict_proba(X_test_best_fold)[:, 1] # ROC曲线需要概率值
# 步骤 5.4: 输出混淆矩阵和分类报告
print("\n--- 混淆矩阵 ---")
cm = confusion_matrix(y_test_best_fold, y_pred_best)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['预测为0 (低风险)', '预测为1 (高风险)'], 
            yticklabels=['实际为0 (低风险)', '实际为1 (高风险)'])
plt.title('Confusion Matrix for the Best Fold')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
print("\n--- 分类报告 (Classification Report) ---")
# 'risk_0' 和 'risk_1' 是更有意义的标签名
report = classification_report(y_test_best_fold, y_pred_best, target_names=['Risk 0 (Low)', 'Risk 1 (High)'])
print(report)
# 步骤 5.5: 绘制ROC曲线并计算AUC值
print("\n--- ROC 曲线与 AUC 值 ---")
fpr, tpr, thresholds = roc_curve(y_test_best_fold, y_pred_proba_best)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for the Best Fold')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
print(f"\n最佳折的 AUC 值为: {roc_auc:.4f}")
```

```python
from sklearn.naive_bayes import GaussianNB

# 预处理：与之前相同，对类别特征进行编码
processed_data = data.copy()
label_encoder = LabelEncoder()
processed_data['loan_purpose'] = label_encoder.fit_transform(processed_data['loan_purpose'])
X = processed_data.drop('risk', axis=1)
y = processed_data['risk']

# 划分训练集和测试集 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)


# --- 2. 构建 Pipeline 与交叉验证 ---
print("\n--- 步骤 2: 构建 Pipeline 与执行交叉验证 ---")

# 创建一个包含标准化和高斯朴素贝叶斯模型的 Pipeline
nb_pipeline = Pipeline([
    ('scaler', StandardScaler()),      # 第一步：数据标准化
    ('model', GaussianNB())            # 第二步：高斯朴素贝叶斯分类器
])

# 设置交叉验证策略（例如，10折分层交叉验证）
cv_strategy = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# 在训练集上执行交叉验证，评估模型的 F1 分数
cv_scores = cross_val_score(
    estimator=nb_pipeline,
    X=X_train,
    y=y_train,
    cv=cv_strategy,
    scoring='f1', # 我们关心的评估指标是 F1
    n_jobs=-1
)

# 计算并输出交叉验证的平均 F1 分数
mean_f1_cv = np.mean(cv_scores)
std_f1_cv = np.std(cv_scores)

print(f"交叉验证各折的 F1 分数: \n{cv_scores}")
print(f"\n交叉验证平均 F1 分数: {mean_f1_cv:.4f} ± {std_f1_cv:.4f}")


# --- 3. 训练模型并预测测试集 ---
print("\n--- 步骤 3: 在完整训练集上训练模型，并预测测试集 ---")

# 在整个训练集上训练 Pipeline
nb_pipeline.fit(X_train, y_train)
print("模型训练完成。")

# 对测试集进行预测
y_pred = nb_pipeline.predict(X_test)

# 获取预测为正类（Risk=1）的概率，用于绘制 ROC 曲线
y_pred_proba = nb_pipeline.predict_proba(X_test)[:, 1]


# --- 4. 评估模型性能 ---
print("\n--- 步骤 4: 评估模型在测试集上的性能 ---")

# 4.1 输出混淆矩阵
print("\n--- 混淆矩阵 ---")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', 
            xticklabels=['预测为0 (低风险)', '预测为1 (高风险)'], 
            yticklabels=['实际为0 (低风险)', '实际为1 (高风险)'])
plt.title('Gaussian Naive Bayes - Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()

# 4.2 输出分类报告
print("\n--- 分类报告 (Classification Report) ---")
report = classification_report(y_test, y_pred, target_names=['Risk 0 (Low)', 'Risk 1 (High)'])
print(report)

# 4.3 绘制 ROC 曲线并计算 AUC 值
print("\n--- ROC 曲线与 AUC 值 ---")
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - GaussianNB')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"\n模型在测试集上的 AUC 值为: {roc_auc:.4f}")
```

### 结果对比分析
| 比较维度 | 决策树模型 | 朴素贝叶斯模型 | 结论 |
|:--|:--|:--|:--|
| 准确率 |    85% |   76.7% | 决策树 |
| 精确率 |   73.1% |   72% | 决策树 |
| 召回率 |   90.5% |   56.3% | 决策树 |
| F1 值 |   80.9% |   63.2% | 决策树 |
| AUC |   0.9261 |   0.8497 | 决策树 |

### 总结
根据各项性能指标，决策树模型的表现全面优于朴素贝叶斯模型。决策树在准确率（85% vs 76.7%）、F1值（80.9% vs 63.2%）和AUC（0.9261 vs 0.8497）等综合性指标上均取得了显著领先。

这种性能差异可能源于两个模型的基本假设。朴素贝叶斯模型的核心是“特征条件独立性假设”，即假定所有特征（如收入、年龄、贷款目的）之间相互独立。然而在真实的信贷场景中，这些特征很可能存在复杂的关联，这个强假设限制了模型的表达能力。相比之下，决策树模型能更好地捕捉特征间的非线性关系和交互作用。

作为风险管理人员，我会选择决策树模型。原因在于其高达90.5%的召回率，远超朴素贝叶斯的56.3%。在信贷风控中，漏掉一个高风险客户可能导致实质性的资金损失，其后果远比错误地拒绝一个低风险客户严重。
