### 课堂作业 分类数据，汇总然后画饼图

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

house = pd.read_csv('../datasets/house_synth.csv',encoding="utf-8")
print(house.head())

h = house["户型"].value_counts()  #value_counts() 用于快速统计一个数据列中每个唯一值出现了多少次,返回的是字典一样的（索引+值）
print(h)

# 设置中文字体，防止乱码
plt.rcParams['font.sans-serif'] = ['Heiti TC']      # 或 ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False        # 解决负号显示问题


labels = h.index
sizes  = h.values
explode = np.zeros_like(h.index)

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
        startangle=140, shadow=True)
plt.title('饼图示例'); plt.axis('equal')  # 保持为圆
plt.show()
```

### 13题 画折线图（plot）

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体，防止乱码
plt.rcParams['font.sans-serif'] = ['Heiti TC']      # 或 ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False        # 解决负号显示问题

x = np.linspace(-10, 10, 400)  #生成400个点 
y = x**2

plt.plot(x, y)

plt.title('y = x² 函数图像')
plt.xlabel('x 轴')
plt.ylabel('y 轴')
plt.grid(True)

plt.show()
```

### 14题 画散点图（scatter，颜色条）

```python
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100

sizes = np.random.randint(10, 201, size=50)

scatter = plt.scatter(x, y, c=y, s=sizes, alpha=0.7, cmap='plasma')

plt.title('随机散点图')
plt.xlabel('X 坐标')
plt.ylabel('Y 坐标')

plt.colorbar(label='y值大小')

plt.show()
```

### 练习15 画柱状图(bar,文本标记)

```python
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
scores = [85, 90, 78, 92, 88]

plt.bar(students, scores, color='skyblue')

m = max(scores)  
m_index = scores.index(m)

#文本标记注释
#   x: 文本的 x 坐标。这里我们用最高分学生的索引 max_index。
#   y: 文本的 y 坐标。我们把它放在柱子顶端稍微往上一点的位置 (max_score + 0.5)，避免重叠。
#   s: 要显示的文本内容。
#   ha: 'center' (horizontal alignment)，让文本水平居中对齐。
#   va: 'bottom' (vertical alignment)，让文本的底部与指定的 y 坐标对齐。
plt.text(m_index, m + 0.5, "max_score",ha='center',va='bottom',fontsize=12, color='red')  

plt.title('考试成绩', fontsize=16)
plt.xlabel('学生', fontsize=12)
plt.ylabel('成绩', fontsize=12)

plt.ylim(0, m + 10) # 设置 Y 轴的范围，让顶部的标注有足够的空间显示 我们将 Y 轴上限设置为比最高分高一点（比如高10分）

plt.show()
```

### 饼图（pie）

```python
labels = ['电子产品', '服装', '食品', '家居用品', '其他']
sizes  = [3500, 4200, 2800, 3100, 1900]

m = max(sizes)
m_index = sizes.index(m)
explode = [0]*len(sizes)
explode[m_index] = 0.1

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
        startangle=140, shadow=True)
plt.title('饼图示例')
plt.axis('equal')  # 保持为圆
plt.show()
```
