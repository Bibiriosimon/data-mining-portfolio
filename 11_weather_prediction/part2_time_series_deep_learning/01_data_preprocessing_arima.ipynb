### 3.2 缺失值检测

```python
# ============================================================
# 3.3 异常值检测与处理
# ============================================================

# ---------- Step 1：物理规则检测 ----------
# 依据成都气候特征设定各变量合理区间
physical_bounds = {
    'MaxTemp':   (-5,  42),
    'MinTemp':   (-10, 35),
    'MeanTemp':  (-8,  38),
    'Precip':    (0,   200),
    'WindSpeed': (0,   60),
    'Radiation': (0,   35)
}

print("=" * 50)
print("Step 1：物理规则异常值检测")
print("=" * 50)

df_clean = df.copy()
physical_count = {}

for col, (lower, upper) in physical_bounds.items():
    if col not in df_clean.columns:
        continue
    mask = (df_clean[col] < lower) | (df_clean[col] > upper)
    count = mask.sum()
    physical_count[col] = count
    if count > 0:
        print(f"{col}：检测到 {count} 条异常值（超出范围 [{lower}, {upper}]）")
        print(f"  异常值详情：\n{df_clean.loc[mask, col]}\n")
    else:
        print(f"{col}：无物理规则异常值")
    df_clean.loc[mask, col] = np.nan

print(f"\n物理规则处理后新增缺失值总计：{sum(physical_count.values())} 条")
```

### 3.3 异常值检测与处理

采用两步异常检测策略：
1. **物理规则检测**：依据成都气候特征设定各变量合理区间（如 MeanTemp ∈ [-8, 38]°C），标记越界值
2. **3×IQR 统计检测**：对通过物理规则的数据进一步用统计方法筛查极端值

**检测结果**：气温类变量无异常；降水量检出 163 条异常值，风速检出 15 条，共 178 条异常值置为 NaN 待插值填补

```python
# ---------- Step 2：IQR统计异常值检测 ----------
print()
print("=" * 50)
print("Step 2：IQR统计异常值检测（3倍IQR）")
print("=" * 50)

iqr_count = {}

for col in df_clean.columns:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_iqr = Q1 - 3.0 * IQR
    upper_iqr = Q3 + 3.0 * IQR

    mask = (df_clean[col] < lower_iqr) | (df_clean[col] > upper_iqr)
    count = mask.sum()
    iqr_count[col] = count

    if count > 0:
        print(f"{col}：检测到 {count} 条统计异常值")
        print(f"  IQR范围：[{lower_iqr:.2f}, {upper_iqr:.2f}]")
        print(f"  异常值：\n{df_clean.loc[mask, col]}\n")
    else:
        print(f"{col}：无IQR统计异常值")

    df_clean.loc[mask, col] = np.nan

print(f"\nIQR检测后新增缺失值总计：{sum(iqr_count.values())} 条")
```

```python
# ---------- 异常值处理汇总 ----------
print()
print("=" * 50)
print("异常值处理汇总")
print("=" * 50)

summary = pd.DataFrame({
    '物理规则异常': physical_count,
    'IQR统计异常':  iqr_count
})
summary['合计'] = summary.sum(axis=1)
print(summary)
print(f"\n全部异常值已置为NaN，待下一步插值填补")
```

```python
# ============================================================
# 3.4 缺失值填补（线性插值）
# ============================================================

print()
print("=" * 50)
print("插值前缺失值统计")
print("=" * 50)
print(df_clean.isnull().sum())

# 线性插值
df_clean = df_clean.interpolate(method='linear')

# 处理首尾可能残余的NaN（前向/后向填充兜底）
df_clean = df_clean.fillna(method='ffill').fillna(method='bfill')

print()
print("插值后缺失值统计")
print("=" * 50)
print(df_clean.isnull().sum())
print("\n所有缺失值已填补完毕")
```

```python
# ============================================================
# ARIMA 建模 第一阶段：序列检验与识别
# 包含：数据准备 → ADF检验 → Ljung-Box检验 → ACF/PACF分析
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
import matplotlib
warnings.filterwarnings('ignore')

matplotlib.rcParams['font.sans-serif'] = ['Heiti TC']
matplotlib.rcParams['axes.unicode_minus'] = False
# ============================================================
# 第一步：数据准备与切分
# ============================================================

print("=" * 60)
print("第一步：数据准备与切分")
print("=" * 60)

# 读取数据
df = pd.read_csv('/Users/xieruixuan/Desktop/时间序列期末论文/chengdu_weather_2018_2023清洗后的.csv', parse_dates=['Date'], index_col='Date')

# 提取目标变量
series = df['MeanTemp'].copy()

# 数据切分
train = series['2019-01-01':'2023-12-31']
test  = series['2024-01-01':'2024-12-31']

print(f"训练集：{train.index[0].date()} ~ {train.index[-1].date()}，共 {len(train)} 天")
print(f"测试集：{test.index[0].date()}  ~ {test.index[-1].date()}，共 {len(test)} 天")
print(f"训练集均值：{train.mean():.2f}°C，标准差：{train.std():.2f}°C")
print(f"测试集均值：{test.mean():.2f}°C，标准差：{test.std():.2f}°C")

# 可视化：原始训练集时序图
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(train.index, train.values, color='steelblue', linewidth=0.8, label='训练集 MeanTemp')
ax.axhline(train.mean(), color='red', linestyle='--', linewidth=1, label=f'均值 {train.mean():.1f}°C')
ax.set_title('训练集日平均气温时序图（2019-2023）', fontsize=13)
ax.set_xlabel('日期')
ax.set_ylabel('气温 (°C)')
ax.legend()
plt.tight_layout()
plt.savefig('arima_step1_raw_series.png', dpi=150)
plt.show()
print("图已保存：arima_step1_raw_series.png\n")


# ============================================================
# 第二步：平稳性检验（ADF检验）
# ============================================================

print("=" * 60)
print("第二步：平稳性检验（ADF检验）")
print("=" * 60)

def run_adf(series, series_name):
    """
    执行ADF检验并打印格式化结果
    H0：序列存在单位根（不平稳）
    p < 0.05 → 拒绝H0 → 序列平稳
    """
    result = adfuller(series.dropna(), autolag='AIC')
    
    adf_stat  = result[0]
    p_value   = result[1]
    n_lags    = result[2]
    n_obs     = result[3]
    crit_vals = result[4]
    
    print(f"\n  [{series_name}]")
    print(f"  ADF统计量    : {adf_stat:.4f}")
    print(f"  p值          : {p_value:.4f}")
    print(f"  使用滞后阶数 : {n_lags}")
    print(f"  有效观测数   : {n_obs}")
    print(f"  临界值：")
    for key, val in crit_vals.items():
        print(f"    {key}: {val:.4f}")
    
    if p_value < 0.05:
        print(f"  ✅ 结论：p={p_value:.4f} < 0.05，拒绝单位根假设 → 序列平稳")
    else:
        print(f"  ❌ 结论：p={p_value:.4f} > 0.05，不能拒绝单位根假设 → 序列不平稳，需要差分")
    
    return p_value < 0.05  # 返回是否平稳

# 对原始序列做ADF检验
stable_raw = run_adf(train, "原始序列 MeanTemp")

# 一阶差分
train_diff1 = train.diff().dropna()
stable_diff1 = run_adf(train_diff1, "一阶差分序列 ΔMeanTemp")

# 根据结果确定 d
if stable_raw:
    d = 0
    print("\n  → 原始序列已平稳，d = 0")
elif stable_diff1:
    d = 1
    print("\n  → 一阶差分后平稳，d = 1")
else:
    train_diff2 = train_diff1.diff().dropna()
    stable_diff2 = run_adf(train_diff2, "二阶差分序列 Δ²MeanTemp")
    d = 2 if stable_diff2 else None
    print(f"\n  → 二阶差分后{'平稳' if stable_diff2 else '仍不平稳'}，d = {d}")

print(f"\n  ✅ 最终确定差分阶数：d = {d}")

# 根据 d 确定后续使用的平稳序列
if d == 0:
    stationary_series = train.copy()
elif d == 1:
    stationary_series = train_diff1.copy()
else:
    stationary_series = train.diff().diff().dropna()

# 可视化：原始序列 vs 差分序列对比
fig, axes = plt.subplots(2, 1, figsize=(14, 7))

axes[0].plot(train.index, train.values, color='steelblue', linewidth=0.8)
axes[0].set_title('原始序列 MeanTemp（不平稳）', fontsize=12)
axes[0].set_ylabel('气温 (°C)')
axes[0].axhline(train.mean(), color='red', linestyle='--', linewidth=1, alpha=0.7)

axes[1].plot(stationary_series.index, stationary_series.values, color='darkorange', linewidth=0.8)
axes[1].set_title(f'{"一阶" if d==1 else "二阶" if d==2 else "原始"}差分序列（平稳）', fontsize=12)
axes[1].set_ylabel('差分值')
axes[1].axhline(0, color='red', linestyle='--', linewidth=1, alpha=0.7)

for ax in axes:
    ax.set_xlabel('日期')
    ax.grid(True, alpha=0.3)

plt.suptitle('ADF检验：差分前后序列对比', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('arima_step2_adf_diff.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n图已保存：arima_step2_adf_diff.png\n")
```

## 第五章 · ARIMA 建模

### 5.1 ARIMA 模型简介

**ARIMA(p, d, q)**（自回归积分移动平均模型）是经典时间序列线性预测框架，由 Box 与 Jenkins 系统化提出。其核心思路：通过对序列进行 d 次差分消除趋势/非平稳性，再以 p 阶自回归项捕捉短期自相关，以 q 阶移动平均项描述随机扰动传导。

**选择 ARIMA 的理由**：日均气温序列具有强短期线性自相关结构（相邻日气温高度相关），与 ARIMA 建模假设高度吻合；且参数少、可解释性强，便于通过残差诊断验证充分性；同时作为基准模型，可量化深度学习方法的实际增益。

**建模流程**：ADF 平稳性检验（确定差分阶数 d）→ Ljung-Box 非白噪声检验 → ACF/PACF 图形分析（初步锁定 p、q 候选范围）→ AIC/BIC 网格搜索（最终定阶）→ 参数估计与残差白噪声检验 → 测试集滚动单步预测

```python
# ============================================================
# 第三步：非白噪声检验（Ljung-Box检验）
# ============================================================

print("=" * 60)
print("第三步：非白噪声检验（Ljung-Box检验）")
print("=" * 60)

print("\n  H0：序列是白噪声（无自相关）")
print("  p < 0.05 → 拒绝H0 → 非白噪声 → 序列有信息可提取 → 可以建模\n")

# 对差分后的平稳序列检验，取多个滞后阶数
lags_to_test = [6, 12, 24, 48]
lb_result = acorr_ljungbox(stationary_series, lags=lags_to_test, return_df=True)

print("  Ljung-Box检验结果：")
print(f"  {'滞后阶数':<10} {'LB统计量':<15} {'p值':<15} {'结论'}")
print("  " + "-" * 55)

all_reject = True
for lag in lags_to_test:
    lb_stat = lb_result.loc[lag, 'lb_stat']
    lb_pval = lb_result.loc[lag, 'lb_pvalue']
    conclusion = "✅ 非白噪声" if lb_pval < 0.05 else "❌ 白噪声"
    if lb_pval >= 0.05:
        all_reject = False
    print(f"  {lag:<10} {lb_stat:<15.4f} {lb_pval:<15.6f} {conclusion}")

print()
if all_reject:
    print("  ✅ 总结：所有滞后阶数下均拒绝白噪声假设")
    print("         平稳序列含有显著自相关结构，可以建立ARIMA模型")
else:
    print("  ⚠️  警告：部分滞后阶数未能拒绝白噪声假设，请检查差分阶数是否合适")

print()


# ============================================================
# 第四步：ACF / PACF 图分析
# ============================================================

print("=" * 60)
print("第四步：ACF / PACF 图分析")
print("=" * 60)

# 设定最大滞后阶数（日数据，看前40阶足够判断短期自相关）
max_lags = 40

# 计算ACF和PACF数值（用于读取具体数值）
acf_vals,  acf_confint  = acf(stationary_series,  nlags=max_lags, alpha=0.05)
pacf_vals, pacf_confint = pacf(stationary_series, nlags=max_lags, alpha=0.05, method='ywm')

# --------------------------------------------------
# 打印ACF/PACF数值，便于定阶判断
# --------------------------------------------------
print(f"\n  平稳序列的ACF值（前{max_lags}阶）：")
print(f"  {'阶数':<6} {'ACF值':<12} {'置信下限':<12} {'置信上限':<12} {'是否显著'}")
print("  " + "-" * 55)

# 95%置信带边界（近似：±1.96/sqrt(n)）
conf_bound = 1.96 / np.sqrt(len(stationary_series))

significant_acf  = []
significant_pacf = []

for i in range(1, max_lags + 1):
    is_sig = abs(acf_vals[i]) > conf_bound
    if is_sig:
        significant_acf.append(i)
    if i <= 15:  # 只打印前15阶
        marker = "★ 显著" if is_sig else ""
        print(f"  {i:<6} {acf_vals[i]:<12.4f} {acf_confint[i][0]-acf_vals[i]:<12.4f} "
              f"{acf_confint[i][1]-acf_vals[i]:<12.4f} {marker}")

print(f"\n  显著的ACF阶数（超出置信带）：{significant_acf}")

print(f"\n  平稳序列的PACF值（前{max_lags}阶）：")
print(f"  {'阶数':<6} {'PACF值':<12} {'是否显著'}")
print("  " + "-" * 40)

for i in range(1, max_lags + 1):
    is_sig = abs(pacf_vals[i]) > conf_bound
    if is_sig:
        significant_pacf.append(i)
    if i <= 15:
        marker = "★ 显著" if is_sig else ""
        print(f"  {i:<6} {pacf_vals[i]:<12.4f} {marker}")

print(f"\n  显著的PACF阶数（超出置信带）：{significant_pacf}")

# --------------------------------------------------
# 初步定阶建议
# --------------------------------------------------
print("\n" + "=" * 60)
print("  初步定阶建议（供信息准则最终确认）")
print("=" * 60)

# 找ACF和PACF截尾位置（连续k阶后首次进入置信带）
def find_cutoff(vals, conf_bound, max_check=10):
    """找截尾位置：第一个|值| < conf_bound 且后续均不显著的阶数"""
    for i in range(1, max_check + 1):
        if abs(vals[i]) < conf_bound:
            return i - 1
    return max_check

acf_cutoff  = find_cutoff(acf_vals,  conf_bound)
pacf_cutoff = find_cutoff(pacf_vals, conf_bound)

print(f"\n  ACF  近似截尾位置：第 {acf_cutoff}  阶后进入置信带 → MA候选 q ≤ {min(acf_cutoff, 3)}")
print(f"  PACF 近似截尾位置：第 {pacf_cutoff} 阶后进入置信带 → AR候选 p ≤ {min(pacf_cutoff, 3)}")

# 给出候选范围
p_max = min(max(pacf_cutoff, 1), 3)
q_max = min(max(acf_cutoff,  1), 3)

print(f"\n  → 建议候选范围：p ∈ {{0, 1, ..., {p_max}}}，q ∈ {{0, 1, ..., {q_max}}}")
print(f"  → d 已确定为 {d}")
print(f"  → 下一步：对所有 (p, {d}, q) 组合计算 AIC / BIC，选最优")

# --------------------------------------------------
# 可视化：ACF / PACF 图
# --------------------------------------------------
fig = plt.figure(figsize=(14, 8))
gs  = gridspec.GridSpec(2, 2, figure=fig)

# --- 左上：平稳序列时序图（局部展示，看波动特征）
ax1 = fig.add_subplot(gs[0, :])  # 占满第一行
ax1.plot(stationary_series.index, stationary_series.values,
         color='steelblue', linewidth=0.8)
ax1.axhline(0, color='red', linestyle='--', linewidth=1, alpha=0.7)
ax1.fill_between(stationary_series.index, stationary_series.values,
                 alpha=0.2, color='steelblue')
ax1.set_title(f'一阶差分后的平稳序列 ΔMeanTemp（d={d}）', fontsize=12)
ax1.set_xlabel('日期')
ax1.set_ylabel('差分值 (°C)')
ax1.grid(True, alpha=0.3)
# --- 左下：ACF 图
ax2 = fig.add_subplot(gs[1, 0])
plot_acf(stationary_series,
         lags=max_lags,
         alpha=0.05,
         ax=ax2,
         color='steelblue',
         vlines_kwargs={'colors': 'steelblue'})
ax2.set_title('ACF（自相关函数）', fontsize=12)
ax2.set_xlabel('滞后阶数')
ax2.set_ylabel('自相关系数')
ax2.axhline( conf_bound, color='red', linestyle=':', linewidth=1.2,
             alpha=0.8, label=f'95%置信带 ±{conf_bound:.3f}')
ax2.axhline(-conf_bound, color='red', linestyle=':', linewidth=1.2, alpha=0.8)
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)
# --- 右下：PACF 图
ax3 = fig.add_subplot(gs[1, 1])
plot_pacf(stationary_series,
          lags=max_lags,
          alpha=0.05,
          ax=ax3,
          method='ywm',
          color='darkorange',
          vlines_kwargs={'colors': 'darkorange'})
ax3.set_title('PACF（偏自相关函数）', fontsize=12)
ax3.set_xlabel('滞后阶数')
ax3.set_ylabel('偏自相关系数')
ax3.axhline( conf_bound, color='red', linestyle=':', linewidth=1.2,
             alpha=0.8, label=f'95%置信带 ±{conf_bound:.3f}')
ax3.axhline(-conf_bound, color='red', linestyle=':', linewidth=1.2, alpha=0.8)
ax3.legend(fontsize=9)
ax3.grid(True, alpha=0.3)
plt.suptitle(f'ARIMA 定阶分析：ACF / PACF 图（差分阶数 d={d}）',
             fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('arima_step4_acf_pacf.png', dpi=150, bbox_inches='tight')
plt.show()
print("图已保存：arima_step4_acf_pacf.png\n")
# --------------------------------------------------
# 打印最终定阶摘要
# --------------------------------------------------
print(f"  候选模型列表：")
print(f"  {'模型':<20} {'待计算指标'}")
print("  " + "-" * 35)
for p in range(p_max + 1):
    for q in range(q_max + 1):
        print(f"  ARIMA({p},{d},{q}){'':10} AIC / BIC")
```

```python
# ==================================================
# 第五步：AIC / BIC 网格搜索定阶
# ==================================================
print("=" * 60)
print("  第五步：候选模型 AIC / BIC 网格搜索")
print("=" * 60)
from itertools import product
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')
# 候选范围
p_candidates = [1, 2, 3, 7]
d_fixed      = 1
q_candidates = [1, 2, 3]
results_list = []
for p, q in product(p_candidates, q_candidates):
    try:
        model = ARIMA(series, order=(p, d_fixed, q))
        res   = model.fit()
        results_list.append({
            'order' : (p, d_fixed, q),
            'AIC'   : round(res.aic, 4),
            'BIC'   : round(res.bic, 4),
            'HQIC'  : round(res.hqic, 4),
            'params': res.params.shape[0]
        })
        print(f"  ARIMA({p},{d_fixed},{q})  AIC={res.aic:.2f}  BIC={res.bic:.2f}  HQIC={res.hqic:.2f}")
    except Exception as e:
        print(f"  ARIMA({p},{d_fixed},{q})  拟合失败：{e}")
# 整理成DataFrame
results_df = pd.DataFrame(results_list)
results_df = results_df.sort_values('AIC').reset_index(drop=True)
print("\n--- 按 AIC 排序 ---")
print(results_df.to_string(index=True))
# 最优模型
best_aic_order = results_df.iloc[0]['order']
best_bic_order = results_df.sort_values('BIC').iloc[0]['order']
print(f"\n  ★ AIC最优模型：ARIMA{best_aic_order}")
print(f"  ★ BIC最优模型：ARIMA{best_bic_order}")
# 若两者不一致，优先选BIC（BIC惩罚更强，模型更简洁）
best_order = best_bic_order
print(f"\n  → 最终选用：ARIMA{best_order}")
```

### 5.2 AIC/BIC 网格搜索定阶

在 ACF/PACF 分析初步确定候选范围 p ∈ {1,2,3,7}、q ∈ {1,2,3} 后，对所有组合逐一拟合并计算 AIC、BIC、HQIC，以 **BIC 最小化**为主要定阶准则（BIC 对参数数量的惩罚更重，倾向于选择更简洁的模型）。

```python
# ==================================================
# 第六步：拟合最优模型与核心参数
# ==================================================
print(f"\n>>> 正在拟合最优模型: ARIMA{best_order} ...")
final_res = ARIMA(series, order=best_order).fit()

# 只保留最核心的参数 p 值对照表，去掉多余的指标
params_summary = pd.DataFrame({
    '系数(Coef)': final_res.params,
    'p值(P>|z|)': final_res.pvalues,
    '显著性': final_res.pvalues.apply(lambda x: '★ 显著' if x < 0.05 else '不显著')
})
print("\n--- 模型参数检验简报 ---")
print(params_summary.round(4))

# ==================================================
# 第七步：残差诊断（精简版）
# ==================================================
print("\n--- 残差诊断报告 ---")
residuals = final_res.resid.dropna()

# 1. 核心白噪声检验：通常只看滞后 10 阶（或12阶）的总体情况即可，不需要打印长表格
from statsmodels.stats.diagnostic import acorr_ljungbox
lb_val = acorr_ljungbox(residuals, lags=[12], return_df=True)
p_value = lb_val['lb_pvalue'].values[0]
print(f"  * 滞后12阶 Ljung-Box 检验 p值 = {p_value:.4f}")
print(f"    结论: {'✅ 残差为白噪声，模型提取信息充分' if p_value > 0.05 else '❌ 残差非白噪声，模型仍可优化'}")

# 2. 一键绘制学术标准的残差诊断四合一图 (替代你原本手工写的几十行画图代码)
fig = final_res.plot_diagnostics(figsize=(12, 8))
plt.suptitle(f'ARIMA{best_order} 残差诊断图汇总', fontsize=12, y=0.98)
plt.tight_layout()
plt.savefig('arima_step7_residual_clean.png', dpi=150)
plt.show()
print("\n诊断图已一键保存至：arima_step7_residual_clean.png")
```

### 5.3 最优模型拟合与残差诊断

**网格搜索结果**：ARIMA(3,1,2) 的 AIC（7501.55）与 BIC（7535.71）在所有候选中均为最低，且参数仅 6 个，不存在明显过参数化问题，最终选定 **ARIMA(3,1,2)**。

**参数显著性**：6 个参数（ar.L1=2.012, ar.L2=-1.253, ar.L3=0.235, ma.L1=-1.903, ma.L2=0.908, σ²=1.790）均在 1% 水平显著（p < 0.001），表明模型结构合理。

**残差诊断**：残差 Ljung-Box 检验（滞后 12 阶）p 值 > 0.05，不能拒绝白噪声假设，说明 ARIMA(3,1,2) 已充分提取序列中的线性自相关信息，无信息遗漏。

```python
# ==================================================
# 第八步：测试集滚动预测评估
# ==================================================
print("=" * 60)
print("  第八步：测试集滚动单步预测评估")
print("=" * 60)

from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

print(f"  训练集：{train.index[0].date()} ~ {train.index[-1].date()}  ({len(train)} 天)")
print(f"  测试集：{test.index[0].date()}  ~ {test.index[-1].date()}   ({len(test)} 天)\n")

# ---------- 滚动单步预测 ----------
history     = list(train.values)
predictions = []

for i in range(len(test)):
    model = ARIMA(history, order=best_order)
    res   = model.fit()
    
    yhat  = res.forecast(steps=1)[0]
    predictions.append(yhat)
    
    history.append(test.iloc[i])
    
    print(f"  [{i+1:03d}/{len(test)}]  "
          f"日期={test.index[i].date()}  "
          f"真实={test.iloc[i]:.2f}°C  "
          f"预测={yhat:.2f}°C  "
          f"误差={abs(test.iloc[i]-yhat):.2f}°C")

# ---------- 计算指标 ----------
actuals     = test.values
predictions = np.array(predictions)

# 1. 导入 r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rmse = np.sqrt(mean_squared_error(actuals, predictions))
mae  = mean_absolute_error(actuals, predictions)
mape = np.mean(np.abs((actuals - predictions) / actuals)) * 100
# 2. 计算 R2
r2   = r2_score(actuals, predictions)

print(f"\n  ┌─────────────────────────────┐")
print(f"  │  测试集滚动预测评估结果      │")
print(f"  │  RMSE  = {rmse:.4f}  °C       │")
print(f"  │  MAE   = {mae:.4f}  °C       │")
print(f"  │  MAPE  = {mape:.4f}  %        │")
print(f"  │  R²    = {r2:.4f}            │") # 3. 打印 R2
print(f"  └─────────────────────────────┘")

# ---------- 可视化 ----------
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# --- 上图：整体时序 + 测试集对比 ---
ax = axes[0]
history_show = train.iloc[-180:]
ax.plot(history_show.index, history_show.values,
        label='历史训练集', color='steelblue', linewidth=0.8, alpha=0.6)
ax.plot(test.index, actuals,
        label='测试集真实值', color='steelblue', linewidth=1.5)
ax.plot(test.index, predictions,
        label='滚动预测值', color='darkorange',
        linewidth=1.5, marker='o', markersize=4)
ax.axvline(test.index[0], color='gray', linestyle='--',
           linewidth=1.2, label='训练/测试分割线')
ax.fill_between(test.index, actuals, predictions,
                alpha=0.2, color='red', label='预测误差')
ax.set_title(f'ARIMA{best_order} 测试集滚动单步预测  RMSE={rmse:.3f}°C  MAE={mae:.3f}°C',
             fontsize=12)
ax.set_ylabel('气温 (°C)')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

# --- 下图：逐日误差柱状图 ---
ax2 = axes[1]
errors = actuals - predictions
colors = ['red' if e < 0 else 'steelblue' for e in errors]
ax2.bar(test.index, errors, color=colors, alpha=0.7, width=0.8)
ax2.axhline(0, color='black', linewidth=0.8)
ax2.axhline(mae, color='darkorange', linestyle='--',
            linewidth=1.2, label=f'+MAE={mae:.2f}°C')
ax2.axhline(-mae, color='darkorange', linestyle='--',
            linewidth=1.2, label=f'-MAE={mae:.2f}°C')
ax2.set_title('逐日预测误差（正=预测偏低，负=预测偏高）', fontsize=12)
ax2.set_xlabel('日期')
ax2.set_ylabel('误差 (°C)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle('ARIMA 测试集滚动预测诊断', fontsize=13)
plt.tight_layout()
plt.savefig('arima_step8_rolling_forecast.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n图已保存：arima_step8_rolling_forecast.png\n")

# ---------- 保存预测结果表 ----------
result_df = pd.DataFrame({
    '日期'    : test.index,
    '真实值'  : actuals.round(2),
    '预测值'  : predictions.round(2),
    '误差'    : (actuals - predictions).round(2),
    '绝对误差': np.abs(actuals - predictions).round(2)
})
result_df = result_df.set_index('日期')
print("--- 逐日预测结果 ---")
print(result_df.to_string())
```
