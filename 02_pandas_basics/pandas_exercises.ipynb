# 练习9

```python
st = pd.read_csv('../datasets/student.csv',encoding="utf-8")
print(st.info())
st.columns=['学生姓名', '语文', '数学', '英语']
print(st.head())
print(st.columns)       #查看所有列名
print(st['语文'])        #查看某一列
print(st.iloc[1:4,:])   #查看某一行用loc或iloc
eng = st["英语"]
print(eng.mean())
```

# 练习10

```python
ch = st['语文']
print(st.loc[(ch>=70)&(eng>=85)])
print(st.loc[:,['语文','数学']])
print(st.iloc[[7,11],[1,3]])    #iloc所有索引必须是数字型的
st['passed'] = st['英语'].apply(lambda x:'Yes' if x>=80 else 'No')
print(st.head())
```

# 练习11

```python
import numpy as np
print('\n语文最大值：',ch.max())
print('\n语文最小值：',ch.min())
math = st['数学']
print('\n数学标准差：',math.std())
passed = st['passed']
print('\n通过人数：',(passed == 'Yes').sum())
print('\n不通过人数：',(passed == 'No').sum())  #count不能解决这个问题，它是对非空的值进行计数
groups = st.groupby('passed')
print(groups.agg(语文平均分=('语文','mean'),数学平均分=('数学','mean'))) #用agg来聚合
```
